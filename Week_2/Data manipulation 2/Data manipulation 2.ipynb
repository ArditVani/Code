```python
import pandas as pd
import numpy as np

sales = pd.read_csv('sales data-set.csv')
features = pd.read_csv('Features data set.csv')
sf = pd.merge(sales, features,  how='left', left_on=['Store','Date','IsHoliday'], right_on = ['Store','Date', 'IsHoliday'])
sf['Date'] = pd.to_datetime(sales['Date'], dayfirst = True)
```

## Cumulatieve statistiek

- Maak een nieuw dataframe sf_2_14_2011 waarin alleen de rijen van afdeling (department) 14 van winkel (store) 2 in 2011 staan.
- Maak een nieuw dataframe sf_ytd_2_14_2011 waarin de rijen uit sf_2_14_2011 zijn gesorteerd op datum (date) in oplopende volgorde.
- YTD is een afkorting van 'Year-to-Date'. Wat betekent dit? Gebruik het Internet als je het antwoord niet weet.
- Voeg de kolom YTD toe waarin de cumulatieve verkopen per week (Weekly Sales) staan.
- Toon de volgende kolommen van sf_ytd_2_14_2011: Date, Weekly Sales en YTD.
- Toon de som van 'Weekly Sales' in sf_ytd_2_14_2011. Komt dit overeen met de de laatste regel van sf_ytd_2_14_2011?

```python
# Create a new data frame sf_2_14_2011 containing only the rows 
# from department 14 of store 2 in 2011.
# Create a new data frame sf_ytd_2_14_2011 in which the rows from sf_2_14_2011 
# are sorted by date (date) in ascending order.

# YTD is an abbreviation of 'Year-to-Date'. What does this mean? Use the 
# Internet if you don't know the answer. 

# Year to date (YTD) refers to the period of 
# time beginning the first day of the current calendar year or 
# fiscal year up to the current date.

# Add the YTD column that shows the cumulative sales per week (Weekly Sales).
#Syntax:  cumsum(axis=None, skipna=True, *args, **kwargs)
#Returns: Cumulative sum of the column

# Display the following columns of sf_ytd_2_14_2011: Date, Weekly Sales and YTD.
# Show the sum of 'Weekly Sales' in sf_ytd_2_14_2011. Does this match the last 
# line of sf_ytd_2_14_2011?

sf_2_14_2011 = sf[(sf["Date"] >= "2011-01-01") & 
                  (sf["Date"] <= "2011-12-31") & 
                  (sf["Store"] == 2 ) & 
                  (sf["Dept"] == 14)]

sf_ytd_2_14_2011 = sf_2_14_2011.sort_values(by = "Date")

sf_ytd_2_14_2011["YTD"] = sf_ytd_2_14_2011["Weekly_Sales"].cumsum()
display(sf_ytd_2_14_2011[["Date","Weekly_Sales", "YTD"]])

total_2011 = sf_2_14_2011["Weekly_Sales"].sum()
display(total_2011)
```

## 4 Ontdubbelen en tellen

```python
# Add the column 'Year' to sf. Use the dt method for this.
#dt.year attribute to return the year of the datetime on e new
#column in this case

sf["Year"] = sf["Date"].dt.year
display(sf.head())
```

```python
# Create the new dataframe stores_year in which the duplicate rows 
# with the same combination of year (year) and store (store) have 
# been removed.
stores_year = sf.drop_duplicates(subset = ["Year", "Store"])
stores_year.head()

# Show the number of stores per year.
stores_year_srt = stores_year[['Year','Store']].sort_values('Year',ascending=False)
stores_year_srt.head()
#count
num_st_year = stores_year["Year"].value_counts()
num_st_year.head()
```

## 7 Draaitabel indicatoren

- Maak een draaitabel sf_piv waarin je de gemiddelde werkloosheid, CPI en benzineprijs per jaar toont.
- Toon ook een regel waarin je de totale gemiddeldes toont.

```python
# Important indicators for a retail chain are unemployment (unemployment),
# Consumer Price Index (CPI) and the petrol price (fuel price). 
# These indicators may have an impact on sales.
# Create a pivot table sf_piv showing the average unemployment, 
# CPI and gasoline price per year.
# Also show a line showing the total averages.

#sf_piv = sf.pivot_tableble(index = "Year", values = ["Unemployment", "CPI", "Fuel_price"], margins = True)

sf_piv = sf.pivot_table(index = 'Year', values = ['Weekly_Sales','Unemployment','CPI','Fuel_Price'], margins = True)
display(sf_piv.head())
```

## 8 Draaitabel temperatuur
- Maak een dataframe sf_2010 waarin je alleen de regels uit 2010 toont.
- Maak de draaitabel sf_2010_piv waarin je alleen de rijen uit sf_2010 opneemt die GEEN vakantieweek (IsHoliday) zijn. De draaitabel toont de gemiddelde temperatuur en de gemiddelde omzet per winkel.
- Sorteer de draaitabel in aflopende volgorde op temperatuur.

```python
# Create a dataframe sf_2010 in which you only show the lines from 2010.
# Create the pivot table sf_2010_piv in which you only include the rows 
# from sf_2010 that are NOT a holiday week (IsHoliday). The pivot table 
# shows the average temperature and average sales per store.
# Sort the pivot table in descending order by temperature.

sf_2010 = sf[sf['Year'] == 2010]
sf_2010_piv=sf_2010[sf_2010['IsHoliday']==False].pivot_table(index='Store', values=['Weekly_Sales', 'Temperature']).sort_values('Temperature', ascending = False)
display(sf_2010_piv.head())
```
