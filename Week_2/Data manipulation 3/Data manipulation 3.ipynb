# Pandas: slicing & indices
- Voer onderstaande cel uit om het dataframe stf aan te maken.

```python
import pandas as pd
import numpy as np
features = pd.read_csv('features.csv', parse_dates = ['Date'], dayfirst = True)
store = pd.read_csv('stores.csv',)
stf = pd.merge(store,features, how='left', left_on='Store', right_on = 'Store')
```

## 1 Inspecteren
- Inspecteer het dataframe met de 3 basisfuncties voor inspecteren.
    - Waar gaat deze dataset over?
    - Welke datatypes zijn er in het dataframe?
    - Wat is de laagste temperatuur in het dataframe?
    - Hoeveel regels heeft het dataframe?

```python
# Inspect the data frame with the 3 basic inspect functions.
# What is this data set about?
# What data types are there in the data frame?
# What is the lowest temperature in the data frame?
# How many lines does the data frame have?

# Descriptive statistics include those that summarize the central tendency, 
# dispersion and shape of a dataset’s distribution, excluding NaN values.
# Analyzes both numeric and object series, as well as DataFrame column sets 
# of mixed data types. The output will vary depending on what is provided. 
# Refer to the notes below for more detail.

display(stf.info, stf.dtypes, stf.describe(), stf.head())

lowest_temp_stf = stf.sort_values(["Temperature"], ascending = True)
display(lowest_temp_stf.head())
print("The lowest temperature is:\n",lowest_temp_stf.iloc[0,4])

```

## 2 Instellen van de index en verwijderen van de index.
- Toon een nieuw dataframe stf_type met "Type" als index.

```python
#Display a new dataframe stf_type with "Type" as index.

stf_type = stf.set_index("Type")
display(stf_type.head())
```

- Verwijder de index zonder de inhoud te verwijderen.

```python
#Delete the index without deleting its contents.

display(stf_type.reset_index())
```

## 3 Filteren
- Maak de list st1_5 met daarin de getallen (integers) 1 tot en met 5.
- Gebruik deze list om een nieuw dataframe stf_1_5 te tonen met daarin alleen de rijen van de winkels 1 tot en met 5.

```python
# Make the list st1_5 containing the numbers (integers) 1 to 5.
# Use this list to display a new dataframe stf_1_5 containing only the rows of stores 1 to 5.
#.isin() function is used to filter the DataFrame rows that contain a list of values. 
#example df2=df[df['Courses'].isin(School)]

st1_5 = [1,2,3,4,5]
stf_1_5 = stf[stf["Store"].isin(st1_5)]
display(stf_1_5.head())
```

## 4 Filteren op index
- Maak het dataframe stf_ind_st met als index 'Store'.
- Toon het dataframe stf_1_5_ind met daarin de rijen van winkels 1 tot en met 5.

```python
# Create the data frame stf_ind_st with index 'Store'.
# Show the data frame stf_1_5_ind containing the rows of stores 1 to 5.

stf_ind_st = stf.set_index("Store")
stf_1_5_ind = stf_ind_st.iloc[[0,1,2,3,4], :]
display(stf_1_5_ind)
```

## 5 Filteren op hierarchische index
- Maak stf_ind_date_store met 'Date' en 'Store' als index. Sorteer de index oplopend.
- Toon stf_2010_02_12_1_5 met daarin de rijen van 12 februari 2010 van de winkels 1 tot en met 5.

```python
# Create stf_ind_date_store with 'Date' and 'Store' as index. Sort the index ascending.
# Display stf_2010_02_12_1_5 containing the February 12, 2010 rows of stores 1 through 5.

stf_ind_date_store = stf.set_index(["Date","Store"]).sort_index(ascending = True)
stf_2010_02_12_1_5 = stf_ind_date_store.loc[("2010-02-12",1):("2010-02-12",5)]
display(stf_2010_02_12_1_5.head())
```

## 6 Filteren op delen van een datum
- Maak stf_ind_date vanuit stf met 'Date' als index. Sorteer de index oplopend.
- Toon stf_2012_1hy met daarin alleen regels uit het eerste half jaar van 2012.

```python
# Create stf_ind_date from stf with 'Date' as index. Sort the index ascending.
# Show stf_2012_1hy containing only lines from the first half of 2012.

stf_ind_date = stf.set_index("Date").sort_index(ascending = True)
stf_2012_1hy = stf_ind_date.loc["2012-01-01":"2012-06-01"]
display(stf_2012_1hy)
```
