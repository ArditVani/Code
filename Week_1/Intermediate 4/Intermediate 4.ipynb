```python
# 1.1 For loops & lists
# Create a for loop that iterates over the elements in the floor list. 
# In this loop, the following occurs:
# the elements (areas) are added up and assigned to the variable total_floor.
# When the loop is finished, total_floor contains the total area of all stores together.
# Determine the number of elements in floor using a function.
# Print the total area, the number of shops and the average area.

total_floor = 0
#start coding here:

for item in floor:
    total_floor = total_floor + item # total_floor += item

print(total_floor)
print(len(floor)) #length floors
print(total_floor / len(floor))  #average area
```

## 5 Integratie
### 5.1 control flow en loops
- Gebruik een for loop om de nieuwe kolom 'Turnover Floor' aan te maken in het storedata dataframe. Toon in deze kolom de omzet (Turnover) per vierkante meter (Floor Space).
- Pas control flow (if, else en/of elif) toe in de loop om de kolom 'Performance' aan te maken:
    - Als de omzet per vierkante meter hoger is dan 210 dan is de performance 'Very Good'.
    - Als de omzet per vierkante meter hoger is dan 200 dan is de performance 'Good'.
    - In alle andere gevallen is de performance 'Poor'.
- Toon de nieuwe kolommen.

```python
# 5 Integration
# 5.1 control flow and loops
# Use a for loop to create the new column 'Turnover Floor' in the storedata dataframe. 
# Show the turnover (Turnover) per square meter (Floor Space) in this column.
# Apply control flow (if, else and/or elif) in the loop to create the 'Performance' column:
# If the turnover per square meter is higher than 210 then the performance is 'Very Good'.
# If the turnover per square meter is higher than 200 then the performance is 'Good'.
# In all other cases the performance is 'Poor'.
# Show the new columns.

# print(storedata.loc[0:"Staff"])
# display(storedata)
for index, row in storedata.iterrows():
    to_fl =row["Turnover"] / row["Floor Space"]
    storedata.loc[index, "Turnover Floor"] = to_fl
    if to_fl > 210:
        storedata.loc[index, "Performance"] = "Very Good"
    elif to_fl > 200:
        storedata.loc[index, "Performance"] = "Very"
    else:
        storedata.loc[index, "Performance"] = "Poor"



storedata[['Turnover Floor','Performance']]
```
