**stamford_staff**

```python
# 2 Control flow
# 2.1 if, else, elif
# stamford_staff
# Print 'big store' if stamford_staff is greater than 8.
# Print 'medium store' if stamford_staff is greater than 6 and less than or equal to 8.
# Print 'small store' in all other cases.
# stamford_loc
# Print stamford_loc if location equals 'Retail Park'.
# If not, print stamford_loc if location equals 'Shopping Centre'.
# Capitalize the first letter of the words in stamford_loc if the previous checks are not true.

stamford_staff = 8
stamford_loc = "shopping centre"

# Start coding here: stamford_staff
if stamford_staff > 8:
    print("big store")
elif stamford_staff > 6:
    print("medium store")
else:
    print("small store")
    
# Start coding here: stamford_loc
if stamford_loc == "Retail Park":
    print(stamford_loc)
elif stamford_loc == "Shopping Centre":
    print(stamford_loc)
else:
    print(stamford_loc.title())
```
