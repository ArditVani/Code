### 2 Samenvoegen van sales and stores
- Gebruik een 'inner join' om sales en stores samen te voegen in een nieuw dataframe sales_stores.
    - Hint: bepaal eerst op welke kolom(men) je de dataframes kunt samenvoegen.
- Laat de eerste 5 rijen zien van sales_stores.
- Laat zien hoeveel rijen sales_stores bevat. Komt dit overeen met wat je verwachtte?

```python
# Use an 'inner join' to merge sales and stores into a new dataframe sales_stores.
# Hint: first determine on which column(s) you can merge the data frames.
# Show the first 5 rows of sales_stores.
# Show how many rows sales_stores contains. Does this match what you expected?

#An inner join requires each row in the two joined dataframes to have matching column values.
#merge on columns with different name
#all we need is to provide the on argument when calling .merge()

sales_stores = sales.merge(stores, on="Store")
display(sales_stores.head())
#display(sales_stores.shape)
sales_store_2 = sales_stores[sales_stores["IsHoliday"]==True]
display(len(sales_store_2.index))
```

### 3 Analyseren samengevoegd dataframe
- Toon de gemiddelde verkopen (Weekly_sales) en de gemiddelde grootte (Size) per winkeltype (Type). Hint: maak gebruik van groupby().

```python
# Show the average sales (Weekly_Sales) and the average size (Size)
# by store type (Type).
# Hint: use groupby().

sales_type = sales_stores.groupby("Type")[["Weekly_Sales", "Size"]].mean()
display(sales_type.head())

sales_type = sales_stores.groupby("Type")[["Weekly_Sales", "Size"]].agg(np.mean)
display(sales_type.head())
```

## 4 Samenvoegen van sales, stores and features 
- Gebruik een 'inner join' om sales, features en stores samen te voegen in een nieuw dataframe sales_stores_features.
    - Hint: bepaal eerst op welke kolom(men) je de dataframes kunt samenvoegen.
- Gebruik 'suffixes' om te laten zien uit welk dataframe een kolom afkomstig is.

```python
# Use an 'inner join' to merge sales, features and stores into one
# new dataframe sales_stores_features.
# Hint: first determine on which column(s) you can merge the data frames. "Store"
# Use 'suffixes' to show from which dataframe a column comes.

sales_stores_features = sales_stores.merge(features, on = ["Store", "Date"],
                                               suffixes = ["_Sales", "_features"])
display(sales_stores_features.head())
```

### 5 Samenvoegen zonder suffixes.
- Er is een mogelijkheid om de 3 dataframes samen te voegen zonder dat er 'suffixes' nodig zijn. Voeg de dataframes samen op een wijze dat er geen suffixes nodig zijn.
    - Hint: je moet hierbij samenvoegen op 3 kolommen.

```python
# There is a possibility to merge the 3 data frames without the need for 'suffixes'. 
# Merge the data frames in such a way that no suffixes are needed.
# Hint: you have to merge on 3 columns.

# sales_stores_features = df1.merge(df2,on='Courses').merge(df3,on='Courses')
# display(sales_stores_features.head())
# display(sales_stores_features.shape)

sales_stores_features = sales_stores.merge(features, on=["Store", "Date", "IsHoliday"])
display(sales_stores_features.head())
```

### 6 Analyseren met een draaitabel
- Maak een draaitabel waarin je de gemiddelde verkopen (Weekly_Sales) laat zien per winkeltype (Type) en IsHoliday. Zorg ervoor dat ook het totale gemiddelde wordt getoond.

```python
# Create a pivot table in which you show the average sales (Weekly_Sales) 
# per store type (Type) and IsHoliday. 
# Make sure that the total average is also shown.

sales_stores_features_piv = pd.pivot_table(sales_stores_features,
                                       values = "Weekly_Sales",
                                        index = ["Type", "IsHoliday"],
                                      aggfunc = {"Weekly_Sales":np.mean},
                                      margins = True)
display(sales_stores_features_piv)
```
