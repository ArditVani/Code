```python
# Bepaal door middel van Python code de toename van de overheidsguitgaven aan onderwijs 
# voor het onderwijstype wetenschappelijk onderwijs in de periode 2000-2010

# Use Python code to determine the increase in government spending on education for science 
# education in the period 2000-2010

cost_2000 = df[(df['Perioden'] == 2000)
            & (df['Type onderwijs'] == 'Tertiair onderwijs|Wetenschappelijk onderwijs')
            & (df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs')]

cost_2010 = df[(df['Perioden'] == 2010)
            & (df['Type onderwijs'] == 'Tertiair onderwijs|Wetenschappelijk onderwijs')
            & (df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs')]


cost_diff = cost_2010['Waarde'].values[0] - cost_2000['Waarde'].values[0]
cost_diff
```

```python
# Selecteer alleen de overheidsuitgaven aan onderwijs in de kolom Type uitgaven sla de data 
# op in een tabel met de naam “overheidsuitgaven"
 
# Select only the government spending on education in the Spending type column save the 
# data in a table called “government spending”
government_spending = df[df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs']
government_spending
```

```python
# Gebruik de nieuwe tabel en maak een draaitabel met in de rijen de Perioden in de kolommen 
# het type onderwijs. Welke manier van aggregeren wordt automatisch gekozen?

# Use the new table and create a pivot table with the periods in the rows and the type of 
# education in the columns. Which way of aggregating is automatically chosen?

# pd.pivot_table(overheidsuitgaven, index='Perioden',columns='Type onderwijs', values='Waarde')
# piv_table = df.pivot_table(columns = "Type onderwijs")
overheidsuitgaven = df[df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs']
overheidsuitgaven
# pandas.pivot_table
# (data, values=None, index=None, columns=None, aggfunc='mean', 
#  fill_value=None, margins=False, dropna=True, margins_name='All', 
#  observed=False, sort=True)[source]
pd.pivot_table(overheidsuitgaven, index='Perioden',columns='Type onderwijs', values='Waarde')
```
